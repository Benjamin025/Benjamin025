# GEOCODING

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
# Intiate the nominatim geocoder
geolocator =Nominatim(user_agent = "monument_geocoder")
```

```python
# Create a rate limiter
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
```

## Creating a new column for address 1

```python
museums["address_1"] = museums["Name"] + ", " + museums["City"] + ", " + museums["Country"]
```

```python
museums
```

### Creating a for loop to iterate through the rows

```python
museums['Full address'] = 0
museums['Latitude'] = 0
museums['Longitude'] = 0

for index, row in museums.iterrows():
  # print(row['address_1'])
  addr = row['address_1']
  location_info = geocode(addr)

  try:

    # saving the address, lat, lon information into the corresponding variables during each iteration
    full_addres = location_info.address
    lat = location_info.latitude
    lon = location_info.longitude

    museums.loc[index, 'Full address' ] = full_addres
    museums.loc[index, 'Latitude' ] = lat
    museums.loc[index, 'Longitude' ] = lon

    print(f'We have gotten the data for {full_addres}')

  except AttributeError:
    museums.loc[index, 'Full address' ] = 'No Info'
    museums.loc[index, 'Latitude' ] = 'No Info'
    museums.loc[index, 'Longitude' ] = 'No Info'
```

## Building Interactive Map

```python
import folium
from folium.plugins import MarkerCluster

```

```python
# Initiate a folium Map
m = folium.Map(loation = [52.359849, 4.885039], zoom_start = 10)
```

```python
for index, row in museums.iterrows():

  # Info to display in popup
  popup_info = f"""
  Name: {row['Name']}<br><br>
  Full Address: {row['Full address']}<br><br>
  Latitude: {row['Latitude']}<br><br>
  Longitude: {row['Longitude']}
  """

  # Creating a marker
  folium.Marker(
      location = [row['Latitude'], row['Longitude']],
      popup = folium.Popup(popup_info, max_width = 300)
  ).add_to(m)
```
