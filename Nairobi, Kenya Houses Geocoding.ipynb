```python
# Display the first few rows to check the current data type of 'price'
print("Before conversion:")
print(df['selling price'].head())
print("Data type:", df['selling price'].dtype)
```

```python
# Remove currency symbols and commas (e.g., 'KSh 130,000,000')
df['selling price'] = df['selling price'].str.replace('KSh', '', regex=False)  # Remove 'KSh'
df['selling price'] = df['selling price'].str.replace(',', '', regex=False)     # Remove commas
```

```python
# Convert the cleaned column to numeric (float or int)
df['selling price'] = pd.to_numeric(df['selling price'], errors='coerce')  # Convert to float
```

```python
# Display the first few rows after conversion
print("\nAfter conversion:")
print(df['selling price'].head())
print("Data type:", df['selling price'].dtype)
```
